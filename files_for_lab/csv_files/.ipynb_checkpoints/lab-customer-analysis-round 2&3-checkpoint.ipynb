**Dealing with the data**

```python
#Read the file as a dataframe

file1 = pd.read_csv('marketing_customer_analysis.csv')
file1 = file1.drop(['Unnamed: 0'],axis=1)
file1
```

**Standardize header names**

**Check and deal with NaN values**

```python
file1.isna().sum()
```

```python
round(file1.isna().sum()/len(file1),4)*100
```

```python
nulls_df = pd.DataFrame(round(file1.isna().sum()/len(file1),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
# Delete NaN

file1 = file1.dropna(subset=['State', 'Response', 'Number of Open Complaints', 'Months Since Last Claim', 'Vehicle Class', 'Vehicle Size'])
file1
```

```python
nulls_df = pd.DataFrame(round(file1.isna().sum()/len(file1),4)*100)
nulls_df
nulls_df = nulls_df.reset_index()
nulls_df
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
#rename NaN Values in the column Vehicle Type 

file1['Vehicle Type'] = file1['Vehicle Type'].fillna("Unknown")
file1
```

**Data time format Extract the months from the dataset and store in a separate column**

```python
file1['Month'] = pd.DatetimeIndex(file1['Effective To Date']).month
file1
file1['Month'] = pd.to_datetime(file1['Effective To Date'], format='%m/%d/%y').dt.strftime('%B')
file1
```

```python
#filtering 

file1_filtered = file1[file1['Month'].isin(['January', 'February', 'March'])]
file1_filtered
```
